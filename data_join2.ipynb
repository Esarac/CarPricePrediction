```python
import pandas as pd
```

```python
df = pd.read_csv('full_data.csv')
model_df = pd.read_csv('model_data.csv', sep=';')

df.head()
```

```python
df['Modelo'] = df['Modelo'].apply(lambda str : str.upper())
model_df['Model'] = model_df['Model'].apply(lambda str : str.upper())
```

```python
# model_df.loc[model_df['Model'] == 'Malibu', 'Category'].iloc[0]
```

```python
models = model_df['Model'].unique()

print(models)
print(len(models))
```

```python
def set_models(prev_model, item):
    if prev_model == item:
        return prev_model
    if len(item) > 2 and item in prev_model:
        print(f"item: {item}")
        print(f"prev model: {prev_model}")
        return item
    else:
        return prev_model
        
print(set_models('cabcdef', 'abcdef'))
print(set_models('aaaaaaaaaaa', 'bbb'))
```

```python
for item in models:
    df['Modelo_abbr'] = df['Modelo'].apply(lambda str : set_models(str, item))
```

```python
def find_category(str):
    try:
        cat = model_df.loc[model_df['Model'] == str, 'Category'].iloc[0]
        
        return cat
    except:
        return 'NA'


df['Category'] = df['Modelo'].apply(lambda str : find_category(str))
```

```python
df['Category'].value_counts()
```

```python
df.loc[df['Modelo'] == 'FRONTIER', 'Category'] = 'Pickup'

df.loc[df['Modelo'] == 'FRONTIER']
```

```python
df['Category'].value_counts()
```

```python
df['Tipo de carroceria'] = df['Tipo de carroceria'].combine_first(df['Category'])
```

```python
df['Tipo de carroceria'].replace('Sed.n', 'Sedan', regex=True, inplace=True)
df['Tipo de carroceria'].replace('SUV', 'Camioneta', regex=True, inplace=True)

# df['Tipo de carroceria'].replace('.*Cami.*', 'Camioneta', regex=True, inplace=True)
# df['Tipo de carroceria'].replace('Hatch.*', 'Hatchback', regex=True, inplace=True)
# df['Tipo de carroceria'].replace('.*Pick.*', 'Camioneta', regex=True, inplace=True)
# df['Tipo de carroceria'].replace('.*Van.*', 'Otro', regex=True, inplace=True)
# df['Tipo de carroceria'].replace('.*g.n.*', 'Otro', regex=True, inplace=True)

df['Tipo de carroceria'].value_counts()
```

```python
grupos_principales = ['Camioneta', 'Sedan', 'Hatchback']

df = df[df['Tipo de carroceria'].isin(grupos_principales)]

df = df[df['Tipo de carroceria'] != 'NA']
```

```python
def obtener_kilometraje(x):
    if (x < 50000):
        return 'BAJO'
    elif (x < 100000):
        return 'MEDIO'
    else:
        return 'ALTO'


df['Kilometraje'] = df['Kilometros'].apply(lambda x : obtener_kilometraje(x))
```

```python
df = df.dropna()
df = df[df['Precio'] < 160000000]
```

```python
print(len(df))
df.to_csv('full_data_test.csv')
```
