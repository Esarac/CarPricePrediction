```python
carroya_df.assign(Titulo = carroya_df['NOMBRE'].fillna('') + ' ' + carroya_df['SUBTITULO'].fillna(''))
```

```python
def obtener_marca(data):
    arr = data.split(' ')
    if arr[0] == 'Mercedes':
        return " ".join(arr[:2])
    else:
        return " ".join(arr[:1])
    
def obtener_modelo(data):
    arr = data.split(' ')
    if arr[0] == 'Mercedes':
        return " ".join(arr[2:])
    else:
        return " ".join(arr[1:])
    

carroya_df['Marca'] = carroya_df['NOMBRE'].apply(lambda str : obtener_marca(str))
carroya_df['Modelo'] = carroya_df['NOMBRE'].apply(lambda str : obtener_modelo(str))
```

```python
tucarro_df.head()
```

```python
carroya_df.head()
```

```python
carroya_df = carroya_df.assign(Titulo = carroya_df['NOMBRE'].fillna('') + ' ' + carroya_df['SUBTITULO'].fillna(''))
carroya_df['Precio'] = carroya_df['PRECIO']
carroya_df['Transmision'] = carroya_df['TIPO DE CAJA']
carroya_df['Ano'] = carroya_df['ANIO']
carroya_df['Tipo de combustible'] = carroya_df['COMBUSTIBLE']
carroya_df['Kilometros'] = carroya_df['KILOMETRAJE']
carroya_df['Img'] = carroya_df['IMAGEN']
carroya_df['Color'] = carroya_df['COLOR']
carroya_df['Puertas'] = carroya_df['PUERTAS']
```

```python
carroya_df.head()
```

```python
carroya_df = carroya_df.drop(['PRECIO', 'PRECIO MENSUAL', 'NOMBRE', 'SUBTITULO', 
                 'KILOMETRAJE', 'ANIO', 'TIPO DE CAJA', 'CILINDRAJE',
                'COMBUSTIBLE', 'COLOR', 'ESTADO', 'UBICACIÓN', 'DIRECCIÓN',
                'PLACA', 'PUERTAS', 'AIRBAGS', 'IMAGEN'], axis=1)
```

```python
df = tucarro_df.append(carroya_df)
```

```python
df['Kilometros'] = df['Kilometros'].astype(str)
df['Kilometros'] = df['Kilometros'].apply(lambda str : str.replace(' km', '').replace('.', ''))

df['Precio'] = df['Precio'].astype(str)
df['Precio'] = df['Precio'].apply(lambda str : str.replace('$', '').replace('.', ''))
```
